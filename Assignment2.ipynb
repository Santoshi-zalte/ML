```python
import numpy as np
from scipy import stats
```

Calculate Z-Score for mathscore column.
Z-Score is also called a standard score.This value/score helps to
understand how far is the data point from the mean.

```python
z = np.abs(stats.zscore(df['Math_Score']))
df
```

Print Z-Score Value. It prints the z-score values of each data item
of the column

```python
print (z)
```

Now to define an outlier threshold value is chosen.

```python
threshold = 0.18
```

Display the sample outliers

```python
sample_outliers = np.where(z <threshold)
sample_outliers
```

IQR

```python
sorted_rscore= sorted(df['Reading_Score'])
```

```python
sorted_rscore
```

Calculate and print Quartile 1 and Quartile 3

```python
q1 = np.percentile(sorted_rscore, 25)
q3 = np.percentile(sorted_rscore, 75)
print(q1,q3)
```

```python
IQR = q3-q1
```

```python
lwr_bound = q1-(1.5*IQR)
upr_bound = q3+(1.5*IQR)
print(lwr_bound, upr_bound)
```

```python
col = ['Reading_Score']
df.boxplot(col)
```

```python
mean=np.mean(sorted_rscore)
mean
```
